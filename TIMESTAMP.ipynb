```python
import pandas as pd

df = pd.read_excel("F:/Payments Reconciliation/Payments Reconciliation/HOLD_TRANSACTIONS_SHARED_FILE.xlsx")

# --- Clean HOLD_TIME ---
df["HOLD_TIME"] = (
    df["HOLD_TIME"]
    .astype(str)
    .str.strip()
    .str.replace(r"\D", "", regex=True)
    .str.zfill(6)
)

# --- Ensure HOLD_DATE column is proper datetime ---
df["DATE"] = pd.to_datetime(df["HOLD_DATE"], errors="coerce")

df["full_timestamp"] = df["DATE"] + pd.to_timedelta(
    df["HOLD_TIME"].str.slice(0,2).astype(int), unit="h"
) + pd.to_timedelta(
    df["HOLD_TIME"].str.slice(2,4).astype(int), unit="m"
) + pd.to_timedelta(
    df["HOLD_TIME"].str.slice(4,6).astype(int), unit="s"
)

print(df[["HOLD_DATE", "HOLD_TIME", "full_timestamp"]].head())

df.to_excel("F:/Payments Reconciliation/Payments Reconciliation/outputs/Hold transactions datetime2.xlsx", index=False)
```
